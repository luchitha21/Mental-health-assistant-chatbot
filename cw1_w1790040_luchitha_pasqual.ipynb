## Techniques Used for chatbots

### K Nearest Neighbor

k-Nearest Neighbors (k-NN) is a simple, yet powerful algorithm for retrieval-based chatbots. It is a non-parametric method that stores all available cases and classifies new cases based on a similarity measure (e.g. distance functions).

In a retrieval-based chatbot, k-NN can be used to find the most similar response to a user's input by comparing it to a dataset of previous inputs and responses. The algorithm would return the k most similar responses, and the chatbot would then choose the most appropriate one to return to the user.

One of the advantages of k-NN is that it is easy to understand and implement. It requires little training data and is relatively insensitive to the choice of distance function. However, it can be computationally expensive when working with large datasets, and the choice of k can have a big impact on the algorithm's performance. and a major downside is in KNN, the prediction phase is slow for a larger dataset. The computation of accurate distances plays a big role in determining the algorithm's accuracy.


### Latent Semantic Analysis

Latent Semantic Analysis (LSA) is a technique that can be used for retrieval-based chatbots by analyzing relationships between a set of documents and the terms they contain. It is a method of natural language processing that aims to identify the underlying meaning or semantics of a text. LSA can be used to identify the most relevant response to a user's input by comparing it to a dataset of previous inputs and responses.

One of the advantages of LSA is that it can handle synonyms and polysemy, which are common in natural language. It can also handle the "synonymy problem" which is a common problem for retrieval-based chatbot, by identifying the underlying meaning of the text and the relationships between words, LSA can find similar meanings even if the words used are different.

LSA works by reducing the dimensionality of the input dataset, which makes it less computationally expensive than other methods such as Neural Network-based retrieval. It also does not require a lot of training data, which can be an advantage for smaller datasets.
However, LSA has some limitations, it can have difficulty handling idiomatic expressions, and it also has difficulties handling new words that have not been seen before. Additionally, LSA can be sensitive to the choice of dimensionality reduction technique and can also be sensitive to the choice of the similarity metric.

Overall, LSA is a powerful technique that can be used for retrieval-based chatbots. It can handle synonyms and polysemy, and it doesn't require a lot of training data. However, it has some limitations, such as difficulty handling idiomatic expressions and new words, and it can be sensitive to the choice of dimensionality reduction technique and similarity metric.


### LSTM (long short-term memory) Network

Long Short-Term Memory (LSTM) networks are a type of RNN that are particularly well-suited for tasks involving sequential data, such as natural language processing. They can learn long-term dependencies in the data, which is useful in a retrievable chatbot context where the goal is to understand the context and intent of a user's message.

In a retrievable chatbot, LSTMs can be used to encode the user's message and compare it to a set of pre-defined responses, or to generate a response based on the user's message and the current context of the conversation.

One of the main advantages of using LSTMs for a retrievable chatbot is their ability to handle sequential data and learn long-term dependencies. This allows the chatbot to understand the context of the conversation and respond in a more natural and relevant way. Additionally, LSTMs can be trained on large amounts of data, which can improve their performance.
However, there are also some disadvantages to using LSTMs in a chatbot context. 
One of the main limitations is that LSTMs can be computationally expensive, which can make it difficult to deploy them in real-time applications. Additionally, LSTMs can be difficult to interpret and debug, which can make it challenging to understand why the chatbot is behaving in a certain way.
Overall, LSTMs can be a powerful tool for building a retrievable chatbot, but it's important to keep in mind their limitations and consider alternative methods as well before taking a final decision.

### Input and chosen technique

## Prototype

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import json
import nltk
from tensorflow import keras
from tensorflow.keras.models import Model
from tensorflow.keras.layers import  Embedding, GlobalMaxPooling1D,LSTM, Dense, MaxPooling1D,Flatten, Input
from tensorflow.keras.preprocessing.text import Tokenizer
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.sequence import pad_sequences
import matplotlib.pyplot as plt
```

### Importing the data

the data is stored in a json file, which can be imported and used as a pandas dataframe, the data was manually processed from a kaggle dataset in order to create a intent and response dataset

```python
with open('mentalhealth.json') as file:
    data = json.load(file)
data
```

```python
#inputting the data to a list
tags =[]
patterns=[]
responses={}
for intent in data['intents']:
    responses[intent['tag']]= intent['responses']
    for lines in intent['patterns']:
        patterns.append(lines)
        tags.append(intent['tag'])
```

```python
#convertion to a dataframe

data = pd.DataFrame({"patterns":patterns,"tags":tags})
data
```

### data pre-processing

pre processing the data and removing the ignore words

```python
import string
data['patterns'] = data['patterns'].apply(lambda wrd:[ltrs.lower() for ltrs in wrd if ltrs not in string.punctuation])
data['patterns'] = data['patterns'].apply(lambda wrd: ''.join(wrd))
data
```

### Tokenization

tokenizing the data to split paragraphs and sentences into smaller units that can be more easily assigned meaning.

```python
tokenizer = Tokenizer(num_words=2000)
tokenizer.fit_on_texts(data['patterns'])
train = tokenizer.texts_to_sequences(data['patterns'])

#apply padding

x_train = pad_sequences(train)

#encoding outputs

le = LabelEncoder()
y_train = le.fit_transform(data['tags'])

input_shape = x_train.shape[1]
input_shape
```

```python
vocabulary = len(tokenizer.word_index)
print("number of uniques words: " ,vocabulary)
output_length = le.classes_.shape[0]
print("output length", output_length)
```

### Neural Network Model

this is a type of recurrent neural network (RNN), specifically, a Long Short-Term Memory (LSTM) network. The LSTM layer allows the model to maintain a memory of previous inputs, which can be useful in cases where the input data has temporal dependencies. The input to the model is passed through an embedding layer, which is followed by the LSTM layer with return sequences set to True and this process is repeated and the output is passed through a Flatten and Dense layer with softmax activation function

```python
i = Input(shape=(input_shape,))
x = Embedding(vocabulary+1,10)(i)
x = LSTM(10, return_sequences=True)(x)
x = Flatten()(x)
x = Dense(output_length,activation="softmax")(x)
model = Model(i,x)
```

```python
opt = keras.optimizers.Adam(learning_rate=0.01)
model.compile(loss="sparse_categorical_crossentropy",optimizer=opt,metrics=['accuracy'])
model.summary()
```

### Train the model

```python

history = model.fit(x_train,y_train,epochs=80)
```

### Plotting model accuracy

```python
plt.plot(history.history['accuracy'], label = 'training set accuracy')
plt.plot(history.history['loss'], label = 'training set loss')
plt.legend()
```

### Testing The model

```python
import random

while True:
    texts_p=[]
    prediction_input = input('You : ')
    
    prediction_input = [letters.lower() for letters in prediction_input if letters not in string.punctuation]
    prediction_input = ''.join(prediction_input)
    texts_p.append(prediction_input)
    
    prediction_input = tokenizer.texts_to_sequences(texts_p)
    prediction_input = np.array(prediction_input).reshape(-1)
    prediction_input = pad_sequences([prediction_input],input_shape)
    
    output =  model.predict(prediction_input)
    output = output.argmax()
    
    response_tag = le.inverse_transform([output])[0]
    print("Joy Bot : ",random.choice(responses[response_tag]))
    if response_tag == "goodbye":
        break
```
